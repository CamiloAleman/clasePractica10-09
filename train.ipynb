```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('train.csv')
```

```python
print(df.head(), '\n')
print(df.info(), '\n')

print(df.describe(), '\n')
```

```python
product_freq = df['Product Name'].value_counts()
print(product_freq.head())
```

```python
top_5_stores = df.groupby('Product ID')['Sales'].sum().nlargest(5)
top_5_stores.plot(kind='bar', title='Top 5 Productos con Mayores Ventas')
plt.ylabel('Ventas Totales')
plt.show()
```

```python
plt.hist(df['Segment'], bins=20, color='skyblue')
plt.title('Distribución de los Segmentos')
plt.xlabel('Segmentos de Ventas')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.bar(df['Category'], df['Sales'], alpha=0.5)
plt.title('Relación entre Categorías de productos y sus ventas')
plt.xlabel('Category')
plt.ylabel('Ventas')
plt.show()
```
